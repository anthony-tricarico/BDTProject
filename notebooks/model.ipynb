# Data Preprocessing

```python
df['departure_time'] = pd.to_timedelta(df['departure_time']).dt.total_seconds()
df['arrival_time'] = pd.to_timedelta(df['arrival_time']).dt.total_seconds()
```

```python
df['time_sin'] = np.sin(2 * np.pi * df['arrival_time'] / 24)
# df['time_cos'] = np.cos(2 * np.pi * df['hours'] / 24)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['encoded_routes'] = le.fit_transform(df['route_short_name'])

print(df['encoded_routes'])
```

We can consider removing the train-test split for model evaluation since we do not need to know how well the generator performs. Ideally, in fact, the generator might be better if it produces data that follows the structure exposed in the prompt but still adds some variability on its own.

```python
from sklearn.model_selection import train_test_split

X = df[['arrival_time', 'stop_id', 'encoded_routes']]
# X = df['stop_id'].to_numpy().reshape(-1,1)
y = df['passengers']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42, shuffle=False) #for panel data it is better to not shuffle around the observations otherwise the temporal structure might get lost
```

# Decision Tree
Only works with the setup having arrival time and stop ID only since they don't need conversion to ordinal variables

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
model = DecisionTreeRegressor()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
from pickle import dump
with open("../generate-passengers/treemodel_out.pkl", "wb") as f:
    dump(model, f, protocol=5)
```
