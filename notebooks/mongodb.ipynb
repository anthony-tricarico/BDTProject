```python
from pymongo import MongoClient
```

```python
# Connect to MongoDB
client = MongoClient("mongodb://localhost:27017/")  # Replace with your URI if needed
db = client["raw"]
collection = db["sensors"]

# Define the aggregation pipeline
pipeline = [
    {
        "$group": {
            "_id": {
                "stop_id": "$stop_id",
                "route": "$route",
                "timestamp": "$timestamp"
            },
            "unique_events": {
                "$addToSet": {
                    "stop_id": "$stop_id",
                    "route": "$route",
                    "timestamp": "$timestamp",
                    "trip_id": "$trip_id",
                    "bus_id": "$bus_id"
                }
            }
        }
    },
    {
        "$project": {
            "stop_id": "$_id.stop_id",
            "route": "$_id.route",
            "timestamp": "$_id.timestamp",
            "passenger_count": { "$size": "$unique_events" }
        }
    }
]



# Run the aggregation
results = list(collection.aggregate(pipeline))

# Optionally, reformat for easier use (e.g., flatten the _id field)
for doc in results:
    doc.update(doc.pop("_id"))

# Example: print the results
for r in results:
    print(r)
```
